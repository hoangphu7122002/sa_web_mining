```python
import pandas as pd

df_train = pd.read_csv("/content/drive/MyDrive/longdoc_sentimentAnalysis/vlsp/vlsp_sentiment_train.csv",index_col=False,on_bad_lines='skip',delimiter='\t')
df_test = pd.read_csv("/content/drive/MyDrive/longdoc_sentimentAnalysis/vlsp/vlsp_sentiment_test.csv",index_col=False,on_bad_lines='skip',delimiter='\t')
```

```python
df_train['text'] = df_train['Data'].apply(lambda x : processor.process_text(x))
df_test['text'] = df_test['Data'].apply(lambda x : processor.process_text(x))
```
